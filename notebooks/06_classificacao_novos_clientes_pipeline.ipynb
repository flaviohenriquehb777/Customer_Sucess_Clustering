# 06 — Classificação de Novos Clientes

Objetivos:
- Demonstrar `pipeline.predict(novo_cliente)`
- Criar `classificar_cliente(novo_df)` que retorna cluster + explicação curta baseada no centróide
- Salvar exemplo de entrada/saída em `reports/`


```python
from pathlib import Path
import sys

sys.path.append(str(Path('..').resolve()))

import json
import pandas as pd

from src.modeling import load_pipeline, predict_with_explanation

PROJECT_ROOT = Path('..').resolve()
REPORTS_DIR = PROJECT_ROOT / 'reports'
REPORTS_DIR.mkdir(parents=True, exist_ok=True)

pipe = load_pipeline(str(REPORTS_DIR / 'kmeans_pipeline.joblib'))
```

## 1) Exemplo de novo cliente


```python
novo_cliente = {
    'log_acessos': 0.0,
    'dias_sem_acessar': 999.0,
    'recorrencia': 0.0,
    'recorrente': 1.0,
    'parcelado': 1.0,
    'tipo_pagamento': 1,
    'nunca_acessou': 1.0,
    'n_transacoes_cliente': 1.0,
    'recencia_compra_dias': 5.0,
    'freq_compra_mensal': 1.0,
    'ativo': 'sim',
    'renovacao': 'nao',
    'faixa_inatividade': '181_plus',
    'metodo_pagamento': 'cartao',
    'parcelas_nao_recorrente_bin': None,
    'status': 'COMPLETE'
}

novo_df = pd.DataFrame([novo_cliente])
novo_df
```

## 2) Função de classificação + explicação


```python
def classificar_cliente(novo_df: pd.DataFrame, top_n: int = 6) -> pd.DataFrame:
    pred = predict_with_explanation(pipe, novo_df, top_n=top_n)
    out = pd.concat([novo_df.reset_index(drop=True), pred], axis=1)
    return out

saida = classificar_cliente(novo_df)
saida
```

## 3) Salvando exemplo de entrada/saída


```python
(REPORTS_DIR / 'novo_cliente_exemplo_entrada.json').write_text(
    json.dumps(novo_cliente, ensure_ascii=False, indent=2),
    encoding='utf-8'
)
saida.to_excel(REPORTS_DIR / 'novo_cliente_exemplo_saida.xlsx', index=False)
saida
```
